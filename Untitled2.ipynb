```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.metrics import classification_report, mean_squared_error,r2_score
```

```python
np.random.seed(42)
data_size = 100
age= np.random.randint(20,70, size = data_size)
bmi = np.random.uniform(18.5,40.0,size=data_size)
blood_pressure =np.random.randint(60,120,size=data_size)
insulin =  np.random.uniform(0,300, size=data_size)
glucose = np.random.uniform(70,200,size=data_size)
```

```python
has_diabetes =(glucose > 140) &(bmi > 25).astype(int)
diabetes_data = pd.DataFrame({'age':age,
                              'BMI':bmi,
                               'blood_pressure':blood_pressure,
                               'Insulin':insulin,
                               'glucose':glucose,
                               'has_diabetes':has_diabetes
                               })
print(diabetes_data.head())
```

```python
print(diabetes_data.info())
```

```python
plt.figure(figsize=(10,6))
plt.scatter(diabetes_data['BMI'],diabetes_data['glucose'], c=diabetes_data['has_diabetes'], cmap='coolwarm',alpha=0.6)
plt.title('BMI VS GULCOSE LEVEL')
plt.xlabel("body mass index(bmi)")
plt.ylabel("gulcose label")
plt.colorbar(label='has_diabetes(1=yes,0=no)')
plt.grid()
plt.show()
```

```python
x_classification = diabetes_data[['age','BMI','blood_pressure','Insulin','glucose']]
y_classification = diabetes_data['has_diabetes']
x_train_class, x_test_class, y_train_class, y_test_class= train_test_split(x_classification,y_classification,test_size=0.2, random_state=42)
knn_classifier=KNeighborsClassifier(n_neighbors=5)
knn_classifier.fit(x_train_class,y_train_class)
```

```python
y_pred_class= knn_classifier.predict(x_test_class)
```

```python
print("/n classification report")
print(classification_report(y_test_class,y_pred_class))
```

```python
x_regression = diabetes_data[['age','BMI','blood_pressure','Insulin']]
y_regression = diabetes_data['glucose']
x_train_reg, x_test_reg, y_train_reg, y_test_reg = train_test_split(x_regression, y_regression, test_size=0.2, random_state=45)
```

```python
knn_regressor = KNeighborsRegressor(n_neighbors=5)
knn_regressor.fit(x_train_reg,y_train_reg)
y_pred_reg = knn_regressor.predict(x_test_reg)
mse = mean_squared_error(y_test_reg,y_pred_reg)
r2= r2_score(y_test_reg,y_pred_reg)
print(f"mean squared error:(mse=2f)")
print("r2 score:",r2)
print(f"r_squared:(r2:2f)")
```

```python
k_values = list(range(1,21))
score =[]
for k in k_values:
    knn_classifier = KNeighborsClassifier(n_neighbors=k)
    knn_classifier.fit(x_train_class,y_train_class)
    score.append(knn_classifier.score(x_test_class,y_test_class))
plt.plot(k_values,score,marker='o')
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.grid()
plt.title("accuracy vs k values")
plt.show()
```
